```python
### do not change the code in this cell
from nltk.corpus import twitter_samples
from nltk.tokenize import word_tokenize
import random,math

def sample_sentences(corpus,sample_size):
    
    size=len(corpus)
    ids=random.sample(range(size),sample_size) #what does this line do?
    sample=[corpus[i] for i in ids]  #this is an example of a list comprehension - what does it do?
    return sample

random.seed(37)

tsample=sample_sentences(twitter_samples.strings(),1000)

twittertokens=[word_tokenize(tweet.lower()) for tweet in tsample]
twittertokens[:5]
```

b)i) For each token in the vocabulary of the tweets stored in `twittertokens`, calculate its document frequency, i.e., the number of tweets that it occurs in, and store this in `docfreq`

[5 marks]

```python
tokens = set()
for tweet in twittertokens:
    for token in tweet:
        tokens.add(token)

tokens
```

```python
import numpy as np

docfreq = {token: value for (token, value) in zip(tokens, np.zeros(len(tokens)))}

for token in tokens:
    for tweet in twittertokens:
        if token in tweet:
            docfreq[token] += 1
            
            
docfreq
```

iii) What is meant if we say that a word is ambiguous with respect to part-of-speech?  Give an example of a word in `tagged_tweets[0]` which could be said to be ambiguous with respect to part-of-speech.  What information does a HMM part-of-speech tagger use to try to resolve the ambiguity?

[6 marks]

A word is said to be ambiguous with respect to part-of-speech tagging if that word could be given multiple pos tags. E.g., read can be a noun or verb. A word is said to be globally ambiguous if its pos tag cannot be derived from context. A word is said to be locally ambiguous if its pos tag can be derived from context. We can see from tagged_tweets[0] that "school" has been tagged as a noun, but it can also be a verb. Its verb meaning is educate. 

A Hidden Markov Model (HMM) is a sequence of observations generated by a sequence of hidden states. HMMs make two assumptions: 1. The current state depends only on the previous state. 2. The current output depends only on the current state.

In relation to pos tagging, the above two assumptions become: 1. The current pos tag depends only on the previous pos tag. 2. The output word depends only on the current pos tag. In the case of pos tagging, the observations of the HMM are the words, and the hidden states are the pos tags. 

In short, a HMM resolves word ambiguity by checking which pos tag a word should have based on the previous word's pos tag.

iv) Find the token with the highest part-of-speech tag ambiguity in the sample.  Explain how you arrived at your answer.
[12 marks]

```python
tagged = []
for tweet in tagged_tweets:
    for pos in tweet:
        tagged.append(pos)
        
tagged
```

```python
def find_tag_distributions(tokentaglist, num=-1):
    tag_dists = {}
    for i, (token, tag) in enumerate(tokentaglist):
        current = tag_dists.get(token, {})
        current[tag] = current.get(tag, 0) + 1
        tag_dists[token] = current
        if num > 0 and i > num:
            print("Max number exceeded")
            break
    return tag_dists
```

```python
dists = find_tag_distributions(tagged)
dists
```

```python
def simple_pos_ambiguity(dists):
    return {word: len(tag_dist.keys()) for word, tag_dist in dists.items()}

ambiguity = simple_pos_ambiguity(dists)
ambiguity
```

```python
print("most ambiguous token: ", max(ambiguity, key=ambiguity.get))
print("number of pos tags: ", max(ambiguity.values()))
```
